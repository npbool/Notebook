```python
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns
```

#univariate, distplot

```python
x=np.random.normal(size=100)
y=np.random.normal(size=200)
sns.distplot(x, hist=False, label='x', vertical=True)
#sns.distplot(y, hist=False, label='y')
plt.legend()
```

#jointplot, JointGrid

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
```

```python
sns.jointplot(x='x', y='y', data=df)
```

```python
x,y = data[:,0], data[:, 1]
```

```python
g = sns.jointplot(x=x,y=y, kind='kde')
g.plot_joint(plt.scatter, c='w', s=30, linewidth=1, marker='+')
g.set_axis_labels('$X$', '$Y$')
```

```python
f,ax = plt.subplots(figsize=(6,6))
```

##pairplot, PairGrid

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris);
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot, kde=False)
g.map_offdiag(sns.kdeplot, n_levels=6)
```

#Conditional dist, hue

```python
tips = sns.load_dataset("tips")
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",
           col="time", row="sex", data=tips);
```

#categorical

```python
sns.stripplot(x="day", y="total_bill", hue='sex', data=tips, jitter=True);
```

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips);
```

```python
sns.boxplot(x="total_bill", y='day',hue="time", data=tips);
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips,
               split=True, inner="stick", palette="Set3");
```

#facet

```python
g = sns.FacetGrid(tips, col="time", hue='sex')
g.map(plt.hist, "tip").add_legend()
```

```python
g = sns.FacetGrid(tips, col="day", row="time")
g.map(sns.barplot, "sex", "total_bill");
```
